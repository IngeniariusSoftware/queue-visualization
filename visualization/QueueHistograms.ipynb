```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
plt.style.use('ggplot')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.serif'] = 'Ubuntu'
plt.rcParams['font.monospace'] = 'Ubuntu Mono'
plt.rcParams['font.size'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.titlesize'] = 12
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['figure.titlesize'] = 12
plt.rcParams['image.cmap'] = 'jet'
plt.rcParams['image.interpolation'] = 'none'
plt.rcParams['figure.figsize'] = (12, 10)
plt.rcParams['axes.grid']=True
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.markersize'] = 8
colors = ['xkcd:pale orange', 'xkcd:sea blue', 'xkcd:pale red', 'xkcd:sage green', 'xkcd:terra cotta', 'xkcd:dull purple', 'xkcd:teal', 'xkcd: goldenrod', 'xkcd:cadet blue',
'xkcd:scarlet']
sns.set_theme(style="darkgrid")
```

```python
figure(figsize=(24, 12), dpi=80)

df = pd.read_csv('dynamic_server_0.csv', delimiter=',')
sns.lineplot(data=df, x='time', y='value');

df = pd.read_csv('dynamic_server_1.csv', delimiter=',')
sns.lineplot(data=df, x='time', y='value');

plt.legend(['Общая очередь', 'Индивидуальная очередь'])
plt.ylabel('Работающий сервер')
plt.xlabel('Время');
```

```python
figure(figsize=(24, 12), dpi=80)

df = pd.read_csv('waiting_times_1.csv')
sns.lineplot(data=df, x=df.index, y='waiting_time');

df = pd.read_csv('waiting_times_0.csv')
sns.lineplot(data=df, x=df.index, y='waiting_time');

plt.legend(['Индивидуальная очередь', 'Общая очередь'])
plt.ylabel('Время ожидания')
plt.xlabel('Время');
```

```python
figure(figsize=(24, 12), dpi=80)

with open('1.txt') as file:
    x1 = [int(line.rstrip()) for line in file]

with open('2.txt') as file:
    x2 = [int(line.rstrip()) for line in file]

zero_x1 = list(filter(lambda x: x == 0, x1))
zero_x2 = list(filter(lambda x: x == 0, x2))
print(len(x1), len(x2))
print(f'{len(zero_x1):_}', f'{len(zero_x2):_}')
print(f'{len(zero_x1) / len(x1) * 100.0:.2f}%', f'{len(zero_x2) / len(x2) * 100.0:.2f}%')
print()

x1 = list(filter(lambda x: x < 300, x1))
x2 = list(filter(lambda x: x < 300, x2))

plt.hist([x1, x2], bins=50)

plt.legend(['Общая очередь', 'Индивидуальная очередь'])
plt.ylabel('Количество')
plt.xlabel('Время ожидания');
```

```python
figure(figsize=(24, 12), dpi=80)

x1 = list(filter(lambda x: x > 0, x1))
x2 = list(filter(lambda x: x > 0, x2))

plt.hist([x1, x2], bins=50)

plt.legend(['Общая очередь', 'Индивидуальная очередь'])
plt.ylabel('Количество')
plt.xlabel('Время ожидания');
```

```python
figure(figsize=(24, 12), dpi=80)

x1 = list(filter(lambda x: x > 0, x1))
x2 = list(filter(lambda x: x > 0, x2))

plt.hist([x1, x2], bins=50, cumulative=True)

plt.legend(['Общая очередь', 'Индивидуальная очередь'])
plt.ylabel('Количество')
plt.xlabel('Время ожидания');
```
